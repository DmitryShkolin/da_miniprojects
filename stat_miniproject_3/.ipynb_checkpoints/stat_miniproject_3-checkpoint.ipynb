```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
data=pd.read_csv('~/london.csv')
```

```python
data.dtypes
```

```python
data['timestamp']=pd.to_datetime(data['timestamp'])
data['weather_code']=data.weather_code.astype(int)
data['is_holiday']=data.is_holiday.astype(int)
data['is_weekend']=data.is_weekend.astype(int)
data['season']=data.season.astype(int)
```

```python
print('Всего {} пропущенных значений.'.format(data.isna().sum().sum()))
print('There are {} empty values.'.format(data.isna().sum().sum()))
```

```python
plt.figure(figsize=(16,9))
sns.lineplot(x='timestamp',y='cnt',data=data)
plt.show()
```

```python
resampled_data=data.set_index('timestamp').resample(rule='D').cnt.sum()\
                .to_frame().reset_index().sort_values('timestamp', ascending=True)
```

```python
plt.figure(figsize=(16,9))
sns.lineplot(x='timestamp',y='cnt',data=resampled_data)
plt.show()
```

```python
resampled_data_rolling=resampled_data.rolling(3, on='timestamp').mean().dropna()
```

```python
resampled_data_rolling[resampled_data_rolling['timestamp']=='2015-07-09']
```

```python
resampled_data_rolling=resampled_data_rolling.\
                        merge(resampled_data,left_on='timestamp',right_on='timestamp',how='left')
```

```python
resampled_data_rolling['difference']=resampled_data_rolling['cnt_x']-resampled_data_rolling['cnt_y']
```

```python
import numpy as np
std=np.std(resampled_data_rolling['difference'])
print(round(std,2))
```

```python
resampled_data_rolling=resampled_data_rolling.rename(columns={'cnt_x':'MA',
                                                             'cnt_y':'cnt'})
resampled_data_rolling['upper_bound']=resampled_data_rolling['MA']+2.576*std
resampled_data_rolling['lower_bound']=resampled_data_rolling['MA']-2.576*std
resampled_data_rolling.sort_values('timestamp', ascending=False).head(1)
```

```python
anomaly_top=resampled_data_rolling[resampled_data_rolling['cnt']>=resampled_data_rolling['upper_bound']]
```

```python
anomaly.sort_values('cnt',ascending=False).head(1)
```

```python
anomaly_bottom=resampled_data_rolling[resampled_data_rolling['cnt']<resampled_data_rolling['lower_bound']]
anomaly_bottom
```
